### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the plots
x = np.linspace(0, 10, 100)  # X values for line chart and scatter plot
y_line = np.sin(x)  # Y values for line chart
categories = ['A', 'B', 'C', 'D']  # Categories for bar chart
values = [5, 7, 3, 8]  # Values for bar chart
scatter_x = np.random.rand(10) * 10  # Random X values for scatter plot
scatter_y = np.random.rand(10) * 10  # Random Y values for scatter plot
pie_labels = ['Category A', 'Category B', 'Category C', 'Category D']
pie_sizes = [15, 30, 45, 10]
pie_colors = ['gold', 'lightcoral', 'lightskyblue', 'yellowgreen']

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# First subplot: Line chart
axs[0, 0].plot(x, y_line, label='y = sin(x)', color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')
axs[0, 0].legend()
axs[0, 0].grid()

# Second subplot: Bar chart
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Third subplot: Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='purple')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')

# Fourth subplot: Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, colors=pie_colors, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and .
# Use different colors for each bar and add a title to the plot.

import matplotlib.pyplot as plt
categories=['A','B','C','D']
values = [5, 7, 3, 9]
colors=['red','blue','black','yellow']

plt.bar(categories,values,color=colors)
plt.title("Bar graph for  the categories and values")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```
